```python
from sklearn.impute import SimpleImputer

# Step 1: Get column names with missing values below index 487(exclude 487)
columns_below_487 = missing_values.index[:487]

# Step 2: Apply SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')  # Using mode to fill missing values
ds[columns_below_487] = imputer.fit_transform(ds[columns_below_487])

# Step 3: Verify that missing values are filled
print(ds[columns_below_487].isna().sum())  # Should show 0 for all columns
```

```python
ds.shape
```

```python
# Initialize an empty list to store the desired column names
columns_with_high_value_count = []

# Iterate through columns starting from index 698
for column in missing_values.index[:487]:
    # Get the value counts for the column
    value_counts = ds[column].value_counts()
    
    # Check if the maximum value count is greater than 96,000
    if not value_counts.empty and value_counts.max() > 96000:
        # Add the column name to the list
        columns_with_high_value_count.append(column)

# Print the resulting list of column names
print(columns_with_high_value_count)
```

```python
# dropping the not useful Columns
```

```python
ds = ds.drop(columns=columns_with_high_value_count)

# Verify the columns are dropped
print(ds.head())
```
